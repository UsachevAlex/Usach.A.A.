```
#* проверка числа
def check(x):
    if isinstance(x, int):
        return x, 1
    return x.a, x.b
#* наименьший делитель
def getNod(a,b):
    while a!=b:
        if a>b:
            a-=b
        else:
            b-=a   
    return a
k=0

#* наш класс рациональных чисел
class Rational:
    def __init__(self, a, b = 1):
        self.a = a
        self.b = b
#*строка
    def __str__(self):
        if self.a==0:
            return "0"
        else:
            return str(self.a) + "/" + str(self.b)
#*сложение 
    def __add__(self, x):
        c, d = check(x)
        return Rational(self.a * d + self.b * c, self.b * d)
#* сложение (правый аргумент)
    def __radd__(self,x):
        c, d = check(x)
        return Rational(self.a * d + self.b * c, self.b * d)
#* вычитание
    def __sub__(self,x):
        c, d = check(x)
        return Rational(self.a * d - self.b * c, self.b * d)
#* обратное вычитание (правый аргумент)
    def __rsub__(self,x):
        c, d = check(x)
        return Rational(self.b * c-self.a * d  , self.b * d)

#* деление
    def __truediv__(self, x):
        c, d = check(x)
        return Rational(self.a * d, self.b * c)
#* обратное деление (правый аргумент)
    def __rtruediv__(self, x):
        c, d = check(x)
        return Rational( self.b * c,self.a * d,)
#* умножение   
    def __mul__(self, x):
        c, d = check(x)
        return Rational(self.a * c, self.b * d)
#* обратное умножение (правый аргумент)
    def __rmul__(self, x):
        c, d = check(x)
        return Rational(self.a * c, self.b * d)
#* равентсво 
    def __eq__(self, x):
        c, d = check(x)
        return self.a * d == self.b * c
#* неравентсво
    def __ne__(self,x):
        c, d = check(x)
        return self.a * d != self.b * c
#*  <
    def __lt__(self,x):
        c, d=check(x)
        return self.a*d<self.b*c
#* <=
    def __le__(self,x):
        c, d=check(x)
        return self.a*d<=self.b*c
#* >
    def __gt__(self,x):
        c, d=check(x)
        return self.a*d>self.b*c
#* >=
    def __ge__(self,x):
        c, d=check(x)
        return self.a*d>=self.b*c
#* степень 
    def __pow__(self,x):
        return Rational(pow(self.a,x),pow(self.b,x))
#* унарноый минуса 
    def __neg__(self):
        return -1*Rational(self.a,self.b)
#* модуль 
    def __abs__(self):
        return Rational(abs(self.a),abs(self.b))
#* +=
    def __iadd__(self, x):
        c,d=check(x)
        return Rational(self.a * d + self.b * c, self.b * d)
#* -=
    def __isub__(self, x):
        c,d=check(x)
        return Rational(self.a * d - self.b * c, self.b * d)
#* /=
    def __itruediv__(self, x):
        c,d=check(x)
        return Rational(self.a * d, self.b * c)
#* *=
    def __imul__(self, x):
        c,d=check(x)
        return Rational(self.a * c, self.b * d)
#* нормализатор
    def normalize(self):
        if self.b<0:
            if self.a>0:
                self.b= abs(self.b)
                self.a= -1*(abs(self.a))
            else:
                self.a= abs(self.a)
                self.b = abs(self.b)
        return Rational(self.a/getNod(self.a,self.b),self.b/getNod(self.a,self.b))

#* десятичный вид 
    def evalf(self):
        return self.a/self.b 
#* целая часть 
    def mix(self):
        while self.a>=self.b:
            global k
            k+=1
            self.a=self.a-self.b
        return k
#*методы для LaTeX
    def toLatex(self):
      a=str(self)










x = (Rational(81) / 19 + 94) * 19
y=Rational(-1,-1)
z=Rational(-1,2)
q=Rational(-1,5)
e=Rational(18,12)
print (x,x==1867)
print (y)
print (Rational.normalize(y))
print (Rational.evalf(z))
print (Rational.mix(e),e)
print (q>=e)
Rational.toLatex(y+z)
    
```
